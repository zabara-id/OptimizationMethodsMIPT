## №1

$f(x)=|2x|$

$f^*(y)=?$

**<u>Реш.</u>** $ \ \phi(x,y)=\langle x, y\rangle-f(x)=xy-|2x|$

$$ \nabla_x\phi(x,y)=
\begin{cases} y-2 \ , if \ x>0& \\
y+2 \ , if \ x<0
\end{cases}\Rightarrow \begin{cases} y=2 \ , if \ x>0 & \\ y=-2 \ , if \ x<0 \end{cases}$$

Тогда:

$$
f^*(y)=\phi(x^*,y)= \begin{cases} 0 \ , if \ x>0 \\ 0 \ , if \ x<0 \end{cases} \Rightarrow \boxed{f^*(y)=0 \ при \ y=\pm2}.
$$

**<u>Ответ:</u>** в рамке.

## №2

Доказать, что если $f(x)=\displaystyle{\inf_{u+v=x}}(g(u)+h(v))$, то $f^*(y)=g^*(y)+h^*(y)$.

**<u>Реш.</u>** $\ f^*(y)=\displaystyle{\sup_{x}}[x^Ty-\displaystyle{\inf_{u+v=x}}(g(u)+h(v))]$

Заметим, что верно 
$$
x^Ty-\displaystyle{\inf_{u+v=x}}(g(u)+h(v))=\displaystyle{\sup_{u+v=x}}[x^Ty-g(u)-h(v)]
$$

Итого, имеем

$$
\displaystyle{\sup_{x}}\displaystyle{\sup_{u+v=x}}[x^Ty-g(u)-h(v)]=\displaystyle{\sup_{u,v}}[(u+v)^Ty-g(u)-h(v)]=\displaystyle{\sup_{u,v}}[u^Ty+v^Ty-g(u)-h(v)]=\displaystyle{\sup_{u}}[u^Ty-g(u)]+\displaystyle{\sup_{v}}[v^Ty-h(v)]=
$$
$$
=g^*(y)+h^*(y), \ ч.т.д.
$$

**<u>Ответ:</u>** доказано.

## №3

$f(x)=ln\displaystyle{\sum_{i=1}^{n}}e^{x_i}$

$f^*(y)=?$

**<u>Реш.</u>** $\phi(x,y)=x^Ty-ln\displaystyle{\sum_{i=1}^{n}}e^{x_i}$

$$
\nabla_x\phi(x,y)=y-\nabla_xln\displaystyle{\sum_{i=1}^{n}}e^{x_i}
$$

$$
\nabla_xln\displaystyle{\sum_{i=1}^{n}}(lny_i)\cdot y_ie^{x_i}=\dfrac{1}{e^{x_1}+...+e^{x^n}}(e^{x_1}, ..., e^{x_n})^T=\dfrac{e^x}{\sum_{i=1}^{n}e^{x_i}}\Rightarrow y_i=\dfrac{e^x_i}{\sum_{i=1}^{n}e^{x_i}}\Rightarrow lny_i=x_i-f(x)
$$

Итого имеем $x_i^*=lny_i+f(x^*)$
$$
\hookrightarrow f^*(y)=\phi(x^*, y)=(lny)^T+f^T(x^*)y-f(x^*)
$$

Здесь $$
f^T(x^*)y=f(x^*)\displaystyle{\sum_{i=1}^{n}}\dfrac{e^x_i}{\sum_{j=1}^{n}e^{x_j}}=f(x^*)\dfrac{\sum_{i=1}^{n}e^{x_i}}{\sum_{j=1}^{n}e^{x_j}}=f(x^*)
$$

Тогда
$$
f^*(y)=(lny)^Ty+f(x^*)-f(x^*)=(lny)^Ty\Rightarrow \boxed{f^*(y)=\displaystyle{\sum_{i=1}^{n}}(lny_i)\cdot y_i}.
$$

$domf^*$ при этом предсталяет собой множество, на котором определены логарифмы под суммой, т.е. $\{ y\in E^* \  | \  y_i>0 \ \forall i\in [1,n] \}$. Иными словами, $domf^*=\mathbb{R}^n_{++}$.

**<u>Ответ:</u>** в рамке.

## №4

Доказать, что если $f(x)=g(Ax)$, то $f^*(y)=g^*(A^{-T}y)$.

**<u>Реш.</u>** Пусть $v=A^{-T}y\Rightarrow y=A^Tv$.

Тогда 
$$
f^*(y)=\displaystyle{\sup_{x}}[y^Tx-f(x)]=\displaystyle{\sup_{x}}[y^Tx-g(Ax)]=\displaystyle{\sup_{x}}[v^TAx-g(Ax)]
$$

Пусть теперь $w=Ax\Rightarrow f^*(y)=\displaystyle{\sup_{w}}[v^Tw-g(w)]=g^*(w)=g^*(A^{-T}y)$, ч.т.д.


**<u>Ответ:</u>** доказано.

## №5

$f(X)=-ln(detX)$, $X\in\mathbb{S}^n_{++}$

$f^*(Y)=?$

**<u>Реш.</u>** $\phi(X,Y)=\langle X,Y \rangle - f(X)\Rightarrow\nabla_{x}\phi(X,Y)=Y+\nabla_x(ln(detX))=Y+X^{-T}$

$Y=-X^{-T}\Rightarrow X^*=-Y^{-1}=(-Y^{-1})^T$, т.к. $X^*\in\mathbb{S}^n_{++}$

Тогда имеем: $f^*(Y)=\phi(X^*, Y)=tr(-Y^{-1}Y)+ln(det(-Y^{-1}))=-tr(I)+ln\dfrac{1}{det(-Y)}\Rightarrow \boxed{f^*(Y)=-n-ln(det(-Y))}$.

Разберёмся с $domf^*$. Матрица $-Y^{-1}$, как видно, положительно определена. Нетрудно догадаться, что $Y$ должны быть отрицательно определены, а также $det(-Y)>0$.

**<u>Ответ:</u>** $f^*(Y)=-n-ln(det(-Y))$.

## №6

$f_{hub}(x)=\begin{cases} \dfrac{1}{2}x^2, \ if \ |x|\leqslant 1 \\ |x|-\dfrac{1}{2}, \ if \ |x|> 1 \end{cases}$

$f_{cshub}(x)=\begin{cases} \dfrac{1}{2}\|x\|_2^2, \ if \ \|x\|_2\leqslant 1 \\ \|x\|_2-\dfrac{1}{2}, \ if \ \|x\|_2 > 1 \end{cases}$

Выпукла ли $f_{cshub}(x)$? 

Найти $f^{*}_{cshub}(y)$.

**<u>Реш.</u>** 

**1)** Выпуклость доказать просто. Совершенно нетрудно видеть, что $f_{hub}(x)$ является выпуклой. Функция $f_{cshub}(x)$ является композицией $f_{hub}(x)$ и $\|\cdot\|_2$, т.е. является композицией двух выпуклых функций. Значит $f_{cshub}(x)-$ выпуклая.

**2)** Найдём $f^*(y)$ на разных участках.

**a)** $\|x\|_2\leqslant 1$ (также как и $\|y\|_2\leqslant1$)

$$f_1(x)=\dfrac{1}{2}\|x\|^2_2\Rightarrow\phi_1(x,y)=\langle x,y \rangle-\dfrac{1}{2}\|x\|^2_2 \ ,$$

$$
\nabla_x\phi_1(x,y)=y-x=0\Rightarrow x^*=y\Rightarrow f_1^*(y)=\phi_1(x^*,y)=y^Ty-\dfrac{1}{2}\|y\|^2_2\Rightarrow  f_1^*(y)=\dfrac{1}{2}\|y\|^2_2 \ .
$$

**б)** $\|x\|_2> 1$ (также как и $\|y\|_2>1$)

$$
f_2(x)=\|x\|_2-\dfrac{1}{2}\Rightarrow f^*_2(y)=\displaystyle{\sup_{x}}[x^Ty-f_2(x)]
$$

$$
\nabla_x\phi_2(x,y)=y-\dfrac{x}{\|x\|_2}\Rightarrow y=\dfrac{x}{\|x\|_2}\Leftrightarrow x=\|x\|_2y
$$

Тогда $$\displaystyle{\sup_{\|x\|_2>1}}\left[\dfrac{x^Tx}{\|x\|_2}-\|x\|_2+\dfrac{1}{2}\right]=\displaystyle{\sup_{\|x\|_2>1}}\left[\|x\|_2y^Ty-\| \|x\|_2y \|_2+\dfrac{1}{2}\right]=\displaystyle{\sup_{\|x\|_2>1}}\left[\|x\|_2\cdot\|y\|_2^2-\|x\|_2\cdot\|y\|_2+\frac{1}{2}\right]=$$
$$
=\displaystyle{\sup_{\|x\|_2>1}}\left[\|x\|_2\cdot\|y\|_2(\|y\|_2-1)+\dfrac{1}{2}\right]
$$

Т.к. $(\|x\|_2)_*=\|x\|_2$, то можно утверждать, что и $\|y\|_2>1$. Тогда видно, что $\displaystyle{\sup_{\|x\|_2>1}}\left[\|x\|_2\cdot\|y\|_2(\|y\|_2-1)+\dfrac{1}{2}\right]=\infty$ на множестве $\|x\|_2>1$, так как при этом $\|y\|_2(\|y\|_2-1)>0$.

По договорённости, $domf^*(y)-$ это такие $y$, что $\displaystyle{\sup_{x}}(\phi(x,y))$ принимает конечные значения. У нас в случае б) получается, что для $\|y\|_2>1$ этот супремум бесконечный $\Rightarrow f^*(y)$ определена только на $\|y\|_2\leqslant1$.

Итого $\boxed{f^*(y)=\dfrac{1}{2}\|y\|^2_2, \ при  \ \|y\|_2\leqslant1}$.

**<u>Ответ:</u>** 1) доказали выпуклость; 2) $f^*(y)=\dfrac{1}{2}\|y\|^2_2$, при $\|y\|_2\leqslant1$.
