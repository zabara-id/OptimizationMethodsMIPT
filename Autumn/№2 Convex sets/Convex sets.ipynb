## №5
Дана дифференцируема функция $f$. Показать, что мн-во направлений строго локального убывания $f$ есть выпуклый конус.

**<u>Реш.</u>**

**1)** Покажем, что заданное мн-во является конусом. $\sqsupset$ мн-во направлений строго локального убывания $f$ в т. $x_0$ есть $A$.

$x\in A\Leftrightarrow \exists \delta > 0: \forall \varepsilon \in (0, \delta)\hookrightarrow f(x_0+\varepsilon x)<f(x_0)$

$\lambda x\in A\Leftrightarrow \exists \delta' = \dfrac{\delta}{\lambda}: \forall \varepsilon \in (0, \delta')\hookrightarrow f(x_0+\lambda\varepsilon x)<f(x_0) -$ также справедливое нер-во для каждго такого $\varepsilon \in \left(0, \dfrac{\delta}{\lambda}\right)\Rightarrow \boxed{A - конус}$. 

**2)** Покажем что $A$ есть выпуклый конус.

$A-$ выпуклый конус $\Rightarrow \forall x_1, x_2 \in A\hookrightarrow \theta_1x_1+\theta_2x_2 \in A $ $ \forall \theta_1,\theta_2\in[0;1]$ и при этом $f(x_0+\varepsilon(\theta_1x_1+\theta_2x_2))<f(x_0)$

Т.к. нас интересуют локальные убывания, то можно считать, что условием принадлежности мн-ву $A$ векторов $x_1$ и $x_2$ является условие

$$
\langle\nabla f^T(x_0), x_1\rangle<0, 
$$

$$
\langle\nabla f^T(x_0), x_2\rangle<0.
$$

$\langle \nabla f^T(x_0), \theta_1x_1+\theta_2x_2\rangle = \theta_1\langle \nabla f^T(x_0), x_1\rangle + \theta_2\langle \nabla f^T(x_0), x_2\rangle < 0$ при условии, что приращение есть, т.е. $\theta_1$ и $\theta_2$ не обращаются в ноль одновременно $\Rightarrow \boxed{A-выпуклый\ конус}$. Надо уточнить, что получающееся множество приращений открыто, ибо для нулевых векторов не определены производные по направлению. Поэтому второй пункт был выполнен в контексте замыкания этого множества.

**<u>Ответ:</u>** доказано.

## №1
$S\subseteq\mathbb{R}^n-$выпуклое

Доказать, что $intS$ и $\overline{S}$ тоже выпуклые.

**<u>Реш.</u>**

**1)** $intS=\{x\in S:\exists \varepsilon >0: B(x, \varepsilon)\subset S\}\subset S-$ множество всех внутренних точек $S$. Из определения внутренности множества очевидно, что если само множество совпадает со своей выпуклой оболочкой, то и внутренность множества совпададает со своей выпуклой оболочкой, ибо в этом случае множество выпуклых комбинаций просто сокращается множеством
граничных точек, при этом принадлежность выпуклых комбинаций оставшихся точек выпуклой оболчки того же множества никак меняться не будет.

**2)** $S-$ выпуклое $\Rightarrow \forall x,y\in S \hookrightarrow \theta x + (1-\theta)y\in S, \ \theta\in[0;1]$ 

$\forall X,Y\in \overline{S}\hookrightarrow\displaystyle{\lim_{x \to \infty}}(x_n)=X\ и \ \displaystyle{\lim_{y \to \infty}}(y_n)=Y\ \forall x_n, y_n\in S$

$\theta X + (1-\theta)Y=\theta\displaystyle{\lim_{x \to \infty}}(x_n)+(1-\theta)\displaystyle{\lim_{y \to \infty}}(y_n)=\displaystyle{\lim_{y \to \infty}}(\theta x_n + (1-\theta)y_n)$

$S-выпуклое\Rightarrow \theta x_n + (1-\theta)y_n\in S$

$\overline{S}$ содержит все предельные точки $S$, т.е. $\displaystyle{\lim_{y \to \infty}}(\theta x_n + (1-\theta)y_n)\in\overline{S}\Rightarrow\theta X + (1-\theta)Y\in\overline{S}\Rightarrow\overline{S}-выпуклое, \ ч.т.д.$

**<u>Ответ:</u>** доказано.

## №2
Показать, что $conv\{xx^T:x\in\mathbb{R}^n, |x|=1\}=\{A\in\mathbb{S}^n_+:tr(A)=1\}$.

**<u>Реш.</u>**

**1)** Пусть $conv\{xx^T:x\in\mathbb(R)^n, |x|=1\}=X$ и $\{A\in\mathbb(S)^n_+:tr(A)=1\}=Y$.

Нужно показать, что $X\subseteq Y\Leftrightarrow\forall x\in X \hookrightarrow x\in Y$. Возьмём $\forall x\in X$.

$A\in\mathbb{S}^n_+:tr(A)=1\Leftrightarrow\langle I, A\rangle=1$

$conv\{xx^T:x\in\mathbb{R}^n, |x|=1\}=\{ \displaystyle{\sum_{i=1}^{n}}\theta_ix_ix^T_i:x_ix^T_i\in\mathbb{S}, \ \displaystyle{\sum_{i=1}^{n}\theta_i=1}, \ \theta_i \geqslant0 \}$

Пусть $x_ix^T_i=B_i$ (обозначение для удобства).

$\displaystyle{\sum_{i=1}^{n}}\theta_iB_i:B_i\in\mathbb{S}, \displaystyle{\sum_{i=1}^{n}}\theta_i=1, \theta_i\geqslant0$

$tr(\theta_1B_1+\theta_2B_2+...+\theta_nB_n)=\theta_1tr(B_1)+\theta_2tr(B_2)+...+\theta_ntr(B_n)$

Заметим, что $tr(B_1)=tr(B_2)=...=tr(B_n)$, т.к. $|x_i|=1, \ \forall i=\overline{1,n}\Rightarrow tr(conv)=(\theta_1+\theta_2+...+\theta_n)\cdot tr(B_1)=tr(x_1x^T_1)=X^2_{11}+X^2_{22}+...+X^2_{nn}=|x|^2=1^2=1$.

**2)** Нужно показать, что $Y\subseteq X\Leftrightarrow\forall y\in Y \hookrightarrow y\in X$.

$A\succeq0\Leftrightarrow \lambda_i\geqslant0, \ \forall i=\overline{1,n}$

$tr(A)=1\Leftrightarrow\displaystyle{\sum_{i=1}^{n}}\lambda_i=1$

$A=Q^T\Lambda Q$, где $Q^TQ=I$ и $\Lambda=diag(\lambda_1, ..., \lambda_n)$

Тот факт, что и внешнее произведение, и матрица $A$ симметричны говорит о том, что последнняя цепочка равентсв из пераого пункта верна в обе стороны и, если учесть спектральное разложение $A$ и то, что $\displaystyle{\sum_{i=1}^{n}}\lambda_i=1=1$, то получим требуемое соотношение.

**<u>Ответ:</u>** доказано.

## №6
$S=\{a\in\mathbb{R}^k:p(0)=1, |p(t)|\leqslant1 \ для \  t\in[\alpha, \beta]$, где $p(t)=a_1+a_2t+...+a_kt^{k-1}$. Выпукло ли данное множество?

**<u>Реш.</u>** Докажем, что выпуклое.

Заметим, что многочлены $p(t)$ однозначно задаются множеством $S$ коэффициентов $a$ как $a^T\overline{t}$, где $\overline{t}-$ вектор степеней $t$. То есть, оперируя многочленами, мы, по сути, оперируем с элементами множества $S$ коэффициентов. Такое возможно благодаря свойствам скалярного произведения. Таким образом, дальнейшие действия обоснованы.

Выберем $p_1=a_1^T\overline{t}$ и $p_2=a_2^T\overline{t}$ (тут $a_1$ и $a_2$ именно **векторы**, т.е. не 1-й и 2-й коэффициенты, а два разных набора всех коэффициентов каждого многочлена). Тут $a_1, a_2\in S$.

$p_1(0)=1, \ |p_1(t)|\leqslant1$ и $p_2(0)=1, \ |p_2(t)|\leqslant1$ для $t\in[\alpha, \beta]$

Пусть $f(t)=\theta p_1(t)+(1-\theta)p_2(t)$, где $\theta\in[0;1]$

$f(0)=\theta p_1(0)+(1-\theta)p_2(0)=\theta + 1 - \theta=1$

$|f(t)|=|\theta p_1(t)+(1-\theta)p_2(t)|\leqslant \theta |p_1(t)|+(1-\theta)|p_2(t)|\leqslant\theta + (1-\theta)\leqslant1$

Итого, $f(t)\in S\Rightarrow \boxed{S-выпуклое}$.

**<u>Ответ:</u>** выпуклое.

## №4
Доказать, что $S=\{x\in\mathbb{R}^2:e^{x_1} < x_2\}$ выпуклое множество.

**<u>Реш.</u>**

**1)** Беру $x\in S$ и $y\in S$ такие, что $e^{x_1}<x_2$ и $e^{y_1}<y_2$.

**2)** Строго $z=\theta x+(1-\theta)y$. Это эквивалентно двум равентсвам $z_1=\theta x_1+(1-\theta)y_1$ и $z_2=\theta x_2+(1-\theta)y_2$.

**3)** Нужно доказать $z\in S:e^{z_1}<z^2$

$e^{z_1}<z^2\Rightarrow z_1<ln(z_2), \ z_2>0$

$\theta x_1 + (1-\theta)y_1<\theta ln(x_2)+(1-\theta) ln(y_2)=ln(x^{\theta}_2y_2^{1-\theta})$

Теперь осталось показать, что $ln(x^{\theta}_2y_2^{1-\theta})<ln(\theta x_2+(1-\theta)y_2)\Leftrightarrow ln\left(\dfrac{x_2^{\theta}y_2^{1-\theta}}{\theta x_2+(1-\theta)y_2}\right)<0\Leftrightarrow x_2^{\theta}y_2^{1-\theta}<\theta x_2+(1-\theta)y_2$

Для этого воспользуемся неравентсвом Юнга: $ab\leqslant \dfrac{a^p}{p}+\dfrac{b^q}{q}, \ где \ \dfrac{1}{p}+\dfrac{1}{q}=1 \ и \ p>1, \ q>1$

В нашем случае $\dfrac{1}{p}=\theta, \ \dfrac{1}{1}=1-\theta, \ a=x_2^{\theta},\  b=y_2^{1-\theta}\Rightarrow x_2^{\theta}b=y_2^{1-\theta}\leqslant\theta x_2+(1-\theta)y_2$.

Итого получили что $\theta x_1 + (1-\theta)y_1\leqslant ln(\theta x_2+(1-\theta)y_2)\Rightarrow \boxed{S - выпуклое, \ ч.т.д.}$

**<u>Ответ:</u>** доказано.

## №3
$\sqsupset K\in\mathbb{R}^n_+-$ конус. Доказать, что если K выпуклый $\Leftrightarrow \{x\in K:\displaystyle{\sum_{i=1}^{n}x_i=1}\} выпукло$.

**<u>Реш.</u>**

**1)** Докажаем слева направо.

По определению $K: \{\forall x\in K\hookrightarrow \lambda x \in K, \ \lambda\geqslant0\}$

$z=\theta_1x_1+\theta_2x_2, \ где \ x_1, x_2\in K, \ \theta_1, \theta_2\geqslant0$

Если K выпуклый, то $\lambda z\in K, \ т.е. \ \boxed{\lambda\theta_1 x_1+\lambda\theta_2 x_2\in K}$. Зпомним это утверждение в рамке.

Нужно показать, что $S=\{x\in K: \displaystyle{\sum_{i=1}^{n}}x_i=1\}$ тоже выпуклое.

Пусть $l=\theta\tilde{x_1}+(1-\theta)\tilde{x_2}$, где $\tilde{x_1}, \tilde{x_2}\in S$, т.е. $\displaystyle{\sum_{i=1}^{n}}\tilde{x_{1i}}=\displaystyle{\sum_{i=1}^{n}}\tilde{x_{2i}}=1$.

Таким образом, чтобы показать, что $S-$ выпуклое множество, то нужно показать, что $\beta l\in K$, где $\beta\geqslant0$

$\beta l=\beta\theta\tilde{x_1}+\beta(1-\theta)\tilde{x_2}$

Из произвольности величин $\theta_1, \theta_2, \lambda$, а также того факта, что элементы исходного конуса содержат в себе элементы множества $S$ (по факту, из конуса в $S$ остались только элементы, удовлетворяющие соотношению $\displaystyle{\sum_{i=1}^{n}x_i=1}\}$), можно сказать, что утверждение в рамке совпадает с требуемым для доказательства выпуклости множества $S$ утверждением, т.е.:

$\lambda\theta_1 x_1+\lambda\theta_2 x_2\in K \thicksim \beta\theta\tilde{x_1}+\beta(1-\theta)\tilde{x_2}\in K$

Осталось показать, что $\displaystyle{\sum_{i=1}^{n}}l_i=1$:

$\displaystyle{\sum_{i=1}^{n}}l_i=\theta\displaystyle{\sum_{i=1}^{n}}\tilde{x_{1i}}+(1-\theta)\displaystyle{\sum_{i=1}^{n}}\tilde{x_{2i}}=\theta+1-\theta=1$

Итого доказали, что $\boxed{S-выпуклое}$.

**2)** Теперь докажем справа налево.

Берём $x,y\in S$. $z=\theta x+(1-\theta)y, \ где \ \theta\in[0;1]$.

$S-выпуклое\Leftrightarrow z\in S, \ т.е. \ \boxed{\theta x+(1-\theta)y\in S}$. Запомним это соотношение в рамке.

$x, y\in K$ значит, что $\lambda x\in K$ и $\kappa y\in K, \ где \ \lambda, \kappa \geqslant0$.

Пусть $u,v\in K$. $K-выпуклый,\  если \ \phi u+\psi v\in K, \ где, \ \phi, \psi\geqslant0$.

Элементы конуса $u$ и $v$ могут быть получены из элементов $x$ и $y$ множества $S$ как, например, $u=\lambda x \ и \ v=\kappa y$. Это можно осуществить путём подбора неотрицательных множителей $\lambda$ и $\kappa$. Действительно, ведь как $\lambda x\in K$, так и $\kappa y\in K$. Суммы $\displaystyle{\sum_{i=1}^{n}}u_i$ и $\displaystyle{\sum_{i=1}^{n}}v_i$ могут быть любыми в рамках ограничений на $\lambda$ и $\kappa$.

Это всё означает, что мы можем написать:
$$
\theta x+(1-\theta)y\thicksim\dfrac{\theta}{\lambda}u+\dfrac{1-\theta}{\kappa}v.
$$

В силу прооизвольности выбора неотрицательных чисел $\lambda$ и $\kappa$, можем положить $\dfrac{\theta}{\lambda}=\phi$ и $\dfrac{1-\theta}{\kappa}=\psi$, где $\phi, \psi \geqslant0$.

Тогда, из всего вышеописанного:  $\theta x+(1-\theta)y\in S\Leftrightarrow\theta x+(1-\theta)y\in K\Leftrightarrow\phi u+\psi v\in K\Rightarrow \boxed{K-выпуклый \ конус}$, ч.т.д.

**<u>Ответ:</u>** доказано.
