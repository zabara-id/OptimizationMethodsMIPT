## №1

$f(x)= \begin{cases} x, \ x>0 \\ ax, \ x\leq0 \end{cases}$

$\partial f = \  ?$

**<u>Реш.</u>** Нетрудно заметитить, что $\boxed{\partial f = \begin{cases} a, \ x<0 \\ [a,1], \ a\leq1, \ x=0 \\ \emptyset, \ a>1, \ x=0 \\  1, \ x>0 \end{cases}}$ по теореме Дубовицкого-Милютина.

При $a>1$ субдифференциал $\partial f(x)$ есть пустое множество, т.к. не будет выполнено $f(x)\geq f(x_0)+\langle g, x-x_0 \rangle$ при $x_0=0$ для любого $x$:
$$
f(x)\geq f(0)+\langle g, x-0 \rangle = 0 + \langle g, x \rangle = \langle g, x \rangle
$$

$$
f(x)\geq \langle g, x \rangle
$$

Пусть $x=1\Rightarrow f(1) \geq g\cdot 1\Rightarrow 1 \geq g$, но в этом случае $g=a>1\Rightarrow $ неравенство для $x=1$ не выполнено, значит субградиент там не определён.

**<u>Ответ:</u>** в рамке.

## №3

$f(x)=\|Ax-b\|_1$

$\partial f(x)= \  ?$

**<u>Реш.</u>** Известно следующее:

$$
\partial \|x\|_1=\{g: \ \|g\|_\infty \leq 1, \ g^Tx=\|x\|_1 \}
$$

$$
\partial (f(Ax+b))(x) = A^T\partial f(Ax+b)
$$

Тогда у нас это даёт: $$\partial ((\|Ax-b\|_1)(x)) = A^T\partial (\|Ax-b\|_1)$$

$$ \partial \|Ax-b\|_1=\begin{cases} B_{\|\cdot\|_*}(0,1), \ Ax=b \\ \{g: \|g\|_\infty\leq1, \ g^T(Ax-b)=\|Ax-b\|_1 \}, \ Ax\neq\end{cases}$$

Итого, $\boxed{\partial f(x)= A^T\cdot \begin{cases} B_{\|\cdot\|_*}(0,1), \ Ax=b \\ \{g: \|g\|_\infty\leq1, \ g^T(Ax-b)=\|Ax-b\|_1 \}, \ Ax\neq b \end{cases}}$. 

Здесь под умножением матрицы на множество подразумевается, что каждый вектор множества в фигурных скобках умножается слева на данную марицу, что даёт нам искомое множество.

**<u>Ответ:</u>** в рамке.

## №4

$f(x)=e^{\|x\|}$

$\partial f(x)= \ ?$

**<u>Реш.</u>** $\partial (e^{\|x\|})=e^{\|x\|}\partial (\|x\|)$

$\partial \|x\| = \begin{cases} B_{\|\cdot\|_*}(0,1), \ x=0 \\ \{s \  : \|s\|_*=1, \ s^Tx=\|x\| \}, \ x\neq0 \end{cases}-$ информация из семинара.

Итого, получаем $\boxed{\partial f = \begin{cases} e^{\|x\|}\cdot B_{\|\cdot\|_*}(0,1), \ x=0 \\ e^{\|x\|}\cdot \{s \  : \|s\|_*=1, \ s^Tx=\|x\| \}, \ x\neq0 \end{cases}}$.

Здесь под умножением числа на множество естественно предполагается, что итоговое множество состоит из векторов множества-множителя, умноженных на число.

**<u>Ответ:</u>** в рамке.

## №5

$f(x)=\dfrac{1}{2}\|Ax-b\|^2_2(x)+\lambda\|x\|_1(x), \ \ \lambda > 0$

$\partial f(x)=\ ?$

**<u>Реш.</u>** По теореме Моро-Рокафеллара $\partial f = \dfrac{1}{2}\partial f_1+\lambda\partial f_2$

По одному из свойств субдифференциала: $\partial \|Ax-b\|^2_2=2\|Ax-b\|_2(x)\partial (\|Ax-b\|_2(x))=2\|Ax-b\|_2(x)_2A^T\partial (\|Ax-b\|_2)$. Так как $(\|\cdot\|_2)_*=\|\cdot\|_2$, то по формуле для субдифференциала произвольной нормы из семинара получаем:
$$
\partial \|Ax-b\|_2=\begin{cases} B_2(0,1), \ Ax=b \\ \{ s: \|s\|_2=1, \ s^T(Ax-b)=\|Ax-b\|_2\}, \ Ax\neq b  \end{cases}
$$

Итого, получим $\partial f_1 = 2\|Ax-b\|_2 A^T \begin{cases} B_2(0,1), \ Ax=b \\ \{ s: \|s\|_2=1, \ s^T(Ax-b)=\|Ax-b\|_2\}, \ Ax\neq b  \end{cases}$

Для второй функции $\partial f_2=\begin{cases} B_{\|\cdot\|_\infty}(0,1), \ x=0 \\ \{ g: \|g\|_\infty \leq 1, \ g^Tx=\|x\|_1 \} , \ x\neq 0 \end{cases}$.

Объединив всё, получим:
$$
\boxed{\partial f(x)= \|Ax-b\|_2 A^T \begin{cases} B_2(0,1), \ Ax=b \\ \{ s: \|s\|_2=1, \ s^T(Ax-b)=\|Ax-b\|_2\}, \ Ax\neq b  \end{cases} + \lambda\cdot\begin{cases} B_{\|\cdot\|_\infty}(0,1), \ x=0 \\ \{ g: \|g\|_\infty \leq 1, \ g^Tx=\|x\|_1 \} , \ x\neq 0 \end{cases}}
$$

Умножения перед множествами тут означают, что все векторы множеств-множителей умножаются на соответствующие числа и матрицы (стоящие слева). Суммы означают суммы Минковского.

**<u>Ответ:</u>** в рамке.
